```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("csv/place.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
sns.scatterplot(x=df['cgpa'],y=df['resume_score'],hue=df['placed'])
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1:]
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
p = Perceptron()
```

```python
lr = LogisticRegression(penalty='l1',solver='liblinear')
```

```python
p1 = Perceptron()
```

```python
p1.fit(X,y)
```

```python
p.fit(X_train,y_train)
```

```python
y_pred = p.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
p.intercept_
```

```python
p.coef_
```

```python
p1.intercept_
```

```python
p1.coef_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values,y.values.ravel(),clf=p,legend=2)
```

```python
plot_decision_regions(X.values,y.values.ravel(),clf=p1,legend=2)
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred1 = lr.predict(X_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
plot_decision_regions(X.values,y.values.ravel(),clf=lr,legend=2)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred1))
```
